# Titanic Survival Prediction

## Problem Statement
The goal of this project is to predict whether a passenger survived the Titanic disaster based on features such as age, gender, passenger class, fare, and family information.


```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
df=pd.read_csv("/content/Titanic-Dataset.csv")
df.head(10)
```

## Data Cleaning

Checking for missing values and removing or filling them.

```python
df.dropna(inplace=True)
```

```python
df=df[['Pclass','Sex','Age','SibSp','Parch','Fare','Embarked','Survived']]
```

```python
x=df.drop("Survived",axis=1)
y=df["Survived"]
```

```python
x["Sex"]=x["Sex"].map({"male":0,"female":1})
```

```python
x["Embarked"]=x["Embarked"].map({"C":1,"S":2,"Q":3})
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x = x.dropna()
```

## Data Visualization

Visualizing survival distribution.

```python
df['Survived'].value_counts().plot(kind='bar')
plt.title("Survival Distribution")
plt.show()
```

## Model Training

Training a Logistic Regression model.

```python
model=LogisticRegression()
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

## Model Evaluation

Evaluating the model using accuracy and confusion matrix.

```python
print("Accuracy:",accuracy_score(y_test,y_pred))
```

```python
print("confusion matrix:",confusion_matrix(y_test,y_pred))
```

## Conclusion

The Logistic Regression model was able to predict Titanic passenger survival with an accuracy of approximately 75%.

Key observations:
- Female passengers had a higher survival rate.
- Passenger class affected survival chances.
- Age and fare also influenced survival probability.
